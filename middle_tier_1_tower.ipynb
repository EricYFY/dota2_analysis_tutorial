```python
from dota2_api import get_api_json

match_id = 3437521264
match_info = get_api_json('https://api.opendota.com/api/matches/{}'.format(match_id))
match_info['objectives']
```

```python
def is_destroy_side_tower_induce_win(match_id):
    match_info = get_api_json('https://api.opendota.com/api/matches/{}'.format(match_id))
    win_camp = 'Radiant' if match_info['radiant_win'] is True else 'Dire'
    if match_info['objectives']:
        for objective in sorted(match_info['objectives'], key=lambda x:x['time']):
            if 'key' in objective and ( 'tower1_top' in str(objective['key']) or 'tower1_bot' in str(objective['key'])):
                tower_camp = 'Dire' if 'goodguys' in objective['key'] else 'Radiant'
                if win_camp == tower_camp:
                    return True
                else:
                    return False
    return None
    
is_destroy_side_tower_induce_win(3437521264)
```

```python
def is_kill_roshan_induce_win(match_id):
    match_info = get_api_json('https://api.opendota.com/api/matches/{}'.format(match_id))
    win_camp = 'Radiant' if match_info['radiant_win'] is True else 'Dire'
    if match_info['objectives']:
        for objective in sorted(match_info['objectives'], key=lambda x:x['time']):
            if 'type' in objective and 'CHAT_MESSAGE_AEGIS' in str(objective['type']):
                roshan_camp = 'Radiant' if objective['slot'] <= 4 else 'Dire'
                if win_camp == roshan_camp:
                    return True
                else:
                    return False
    return None
    
is_kill_roshan_induce_win(3437521264)
```

```python
true_count = 0
match_count = 0

for mi in match_ids:
#     is_induce_win = is_destroy_mid_tower_induce_win(mi['match_id'])
#     is_induce_win = is_kill_roshan_induce_win(mi['match_id'])
    is_induce_win = is_destroy_side_tower_induce_win(mi['match_id'])
    if is_induce_win is not None:
        match_count += 1
        if is_induce_win is True:
            true_count += 1
            print(mi, 'True')
        else:
            print(mi, 'False')
    else:
        print(mi, 'No Data')

print(match_count, true_count)
```
