```python
#获取数据源
from dota2_api import get_api_json

sql = """SELECT
matches.match_id,
matches.start_time,
matches.radiant_win,
leagues.name leaguename,
matches.picks_bans
FROM matches
JOIN leagues using(leagueid)
where leagueid = 9943 and to_timestamp(start_time) >= now() - interval '1 month'
ORDER BY start_time desc"""


rows = get_api_json('https://api.opendota.com/api/explorer?sql={}'.format(sql))['rows']
rows
```

```python
import pickle

with open('cn_heros.pickle', 'rb') as handle:  
    cn_heros = pickle.load(handle)

cn_heros
```

```python
picks = []
for row in rows:
    radiant_picks = [ cn_heros[bp['hero_id']] for bp in row['picks_bans'] if bp['team'] == 0 and bp['is_pick']]
    dire_picks = [ cn_heros[bp['hero_id']] for bp in row['picks_bans'] if bp['team'] == 1 and bp['is_pick']]
    picks.append(radiant_picks)
    picks.append(dire_picks)
picks
```

```python
from pymining import itemmining

relim_input = itemmining.get_relim_input(picks)
item_sets = itemmining.relim(relim_input, min_support=4)

for itemset, support_count in sorted(item_sets.items(), key = lambda x:str(x[0])):
    if len(itemset) > 1:
        print(itemset, support_count)

```
